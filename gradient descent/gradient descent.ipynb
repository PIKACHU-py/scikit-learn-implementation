# Gradient Descent

```python
import numpy as np
```

```python
def gradient_descent(X,y,learning_rates,iterations):
    m_curr=b_curr=0
    iteration=iterations
    n=len(X)
    learning_rate=learning_rates
    for i in range(iteration):
        y_predicted=m_curr*X+b_curr
        cost=(1/2)*sum([val**2 for val in (y-y_predicted)])
        m=-(2/n)*sum(X*(y-y_predicted))
        b=-(2/n)*sum(y-y_predicted)
        m_curr=m_curr-learning_rate*m
        b_curr=b_curr-learning_rate*b
        print('m= {} b ={} cost ={} iteration={}'.format(m_curr,b_curr,cost,i))
    return m_curr,b_curr
        
        
        
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
m,b=gradient_descent(x,y,0.08,1500)
```
